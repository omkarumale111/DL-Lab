```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras import Model
from sklearn.manifold import TSNE

# Load the MNIST dataset
(x_train, _), (x_test, _) = tf.keras.datasets.mnist.load_data()

# Normalize the pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape the images to be flat arrays (28*28 = 784)
x_train = x_train.reshape((len(x_train), 28*28))
x_test = x_test.reshape((len(x_test), 28*28))

print(f"Training data shape: {x_train.shape}")
print(f"Testing data shape: {x_test.shape}")

# Define the input layer
input_dim = 784
input_layer = Input(shape=(input_dim,))

# Define the encoder
encoded = Dense(128, activation='relu')(input_layer)
encoded = Dense(64, activation='relu')(encoded)
# Bottleneck layer
encoded = Dense(32, activation='relu')(encoded)

# Define the decoder
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(decoded)
# Output layer
decoded = Dense(input_dim, activation='sigmoid')(decoded)

# Create the autoencoder model
autoencoder = Model(input_layer, decoded)

# Optionally, create separate encoder and decoder models
encoder = Model(input_layer, encoded)

encoded_input = Input(shape=(32,)) # Shape of the bottleneck layer
decoder_layer = autoencoder.layers[-3](encoded_input) # Use the layers from the trained autoencoder
decoder_layer = autoencoder.layers[-2](decoder_layer)
decoder_layer = autoencoder.layers[-1](decoder_layer)
decoder = Model(encoded_input, decoder_layer)

autoencoder.summary()
encoder.summary()
decoder.summary()

# Compile the autoencoder model
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder model
history = autoencoder.fit(x_train, x_train,
                          epochs=50,
                          batch_size=256,
                          shuffle=True,
                          validation_data=(x_test, x_test))

print("Autoencoder training complete.")

# Predict the reconstructed outputs for the test data
reconstructed_x_test = autoencoder.predict(x_test)

# Calculate the Mean Squared Error (MSE) between original and reconstructed test data
mse = tf.keras.metrics.MeanSquaredError()
mse.update_state(x_test, reconstructed_x_test)
reconstruction_mse = mse.result().numpy()

# Print the calculated MSE
print(f"Reconstruction MSE on test data: {reconstruction_mse}")

# Use the trained encoder to predict the lower-dimensional representation of the training data
encoded_train_data = encoder.predict(x_train)

# Use the trained encoder to predict the lower-dimensional representation of the test data
encoded_test_data = encoder.predict(x_test)

# Print the shapes of the resulting encoded data
print(f"Shape of encoded training data: {encoded_train_data.shape}")
print(f"Shape of encoded test data: {encoded_test_data.shape}")

# Select a few images for visualization
n = 10 # Number of images to display
plt.figure(figsize=(20, 4))
for i in range(n):
    # Display original images
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_title("Original")

    # Display reconstructed images
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(reconstructed_x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_title("Reconstructed")
plt.suptitle("Original vs Reconstructed Images")
plt.show()

# Visualize the lower-dimensional representation using t-SNE if bottleneck dimension > 2
# Assuming bottleneck dimension is 32 from previous steps, which is > 2
print("Visualizing lower-dimensional representation using t-SNE...")
# Reduce the number of samples for t-SNE to make it faster
tsne_samples = 1000
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=300)
encoded_tsne = tsne.fit_transform(encoded_test_data[:tsne_samples])

plt.figure(figsize=(10, 8))
# You would ideally have the original labels (y_test) to color the points by digit
# Since we only loaded x_test, we cannot color by digit without loading y_test as well
# Plotting without color for simplicity based on the available variables
plt.scatter(encoded_tsne[:, 0], encoded_tsne[:, 1], s=5)
plt.title("t-SNE Visualization of Encoded Test Data")
plt.xlabel("t-SNE component 1")
plt.ylabel("t-SNE component 2")
plt.show()
```
